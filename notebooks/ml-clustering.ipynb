```python
# Load NASA's public exoplanet data (NO login needed)
import pandas as pd
import pyvo as vo
service = vo.dal.TAPService("https://exoplanetarchive.ipac.caltech.edu/TAP")
```

```python
resultset = service.search("select * from ps where pl_masse is not null and pl_rade is not null and st_teff is not null and st_rad is not null")
table = resultset.to_table()
df = table.to_pandas()
print(f"Number of rows in the dataset: {df.shape[0]}")
```

```python
dataset = df[['pl_name', 'pl_orbper', 'pl_rade', 'pl_masse', 'st_teff', 'st_rad', 'pl_orbper', 'st_spectype', 'pl_eqt', 'pl_orbeccen', 'pl_orbincl', 'pl_bmassj', 'pl_radj', 'pl_dens', 'pl_insol', 'sy_snum', 'sy_pnum', 'cb_flag']]
dataset
```
